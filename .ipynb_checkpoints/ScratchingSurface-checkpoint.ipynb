```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
import numpy as np 
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn import linear_model
import graphviz 

# features to consider
def mk_features(odf):
    fts = list(odf.columns.values)
    fts.remove('nr-of-gadgets')
    fts.remove('binary-size')
    fts.remove('unit')
    return fts

def mk_hist_size(odf, SYSTEM_NAME):    
    fig = plt.figure()
    ax = odf['binary-size'].plot(kind='hist',bins=50)
    ax.set_xlabel("binary size")
    fig.savefig("output/" + SYSTEM_NAME + '-hist.png')
    plt.show()
    
def mk_hist_gadget(odf, SYSTEM_NAME):
    fig_gadget = plt.figure()
    ax_g = odf['binary-size'].plot(kind='hist',bins=50)
    ax_g.set_xlabel("gadget")
    fig_gadget.savefig("output/" + SYSTEM_NAME + '-gadget.png')
    plt.show()

def process_system(DATASET, SYSTEM_NAME, tsize):     
    odf = pd.read_csv(DATASET)
 
    fts = mk_features(odf)
    mk_hist_size(odf, SYSTEM_NAME)
    odf['binary-size'].plot(kind='box')
    mk_hist_gadget(odf, SYSTEM_NAME)
    odf['nr-of-gadgets'].plot(kind='box')
    corr_size_gadget = odf['binary-size'].corr(odf['nr-of-gadgets'])
    # print("correlation size, gadget:", corr_size_gadget)
    
 
    
    def mean_absolute_percentage_error(y_pred, y_true):
        return np.average(np.abs(y_pred - y_true) / np.abs(y_true))

    def eval_model(target, y_pred, X_test, y_test): 
        print('Mean squared error: %.2f'
              % mean_squared_error(y_test, y_pred))
        # The coefficient of determination: 1 is perfect prediction
        print('Coefficient of determination: %.2f'
              % r2_score(y_test, y_pred))
        
        mape = mean_absolute_percentage_error(y_pred, y_test)
        print("MAPE", mape)

        # Plot outputs
        plt.scatter(X_test.index, y_test,  color='black')
        plt.plot(X_test.index, y_pred, color='blue', linewidth=3)

        plt.xticks(())
        plt.yticks(())

        plt.show()
        return mape

    def mk_linear_regression(target, test_size=0.8):
        regr = linear_model.LinearRegression()
        X_train, X_test, y_train, y_test = train_test_split(odf[fts], odf[target], test_size=test_size, random_state=42)
        regr.fit(X_train, y_train)
        print(regr.coef_)
        coeffs = pd.DataFrame(regr.coef_, fts, columns=['Coefficients'])
        print(coeffs.sort_values(by="Coefficients"))
        coeffs.plot(kind='barh', figsize=(9, 7))
        plt.axvline(x=0, color='.5')
        plt.subplots_adjust(left=.3)
        plt.show()

        y_pred = regr.predict(X_test)
        return eval_model(target, y_pred, X_test, y_test)




    def mk_treepredictor(target, decision_tree, test_size=0.8): 
        if decision_tree:
            clf = tree.DecisionTreeRegressor()
        else:
            clf = RandomForestRegressor() 
        X_train, X_test, y_train, y_test = train_test_split(odf[fts], odf[target], test_size=test_size, random_state=42)

        clf = clf.fit(X_train, y_train)

        if decision_tree:
            dot_data = tree.export_graphviz(clf, out_file=None, feature_names=fts,  
                          filled=True, rounded=True,  
                          special_characters=True)
            graph = graphviz.Source(dot_data) 
            graph.render(SYSTEM_NAME + "/" + SYSTEM_NAME)


        y_pred = clf.predict(X_test)
        

        importance = clf.feature_importances_
        for i, v in enumerate(importance):
            print('Feature: %d, %s, Score: %.5f' % (i, odf.columns[i], v))
        plt.bar([x for x in range(len(importance))], importance)
        plt.show()
        return eval_model(target, y_pred, X_test, y_test)
        
    nrepeat=5
    mape_linear_size = np.mean([mk_linear_regression('binary-size', tsize) for i in range(0, nrepeat)])
    mape_DT_size = np.mean([mk_treepredictor('binary-size', True, tsize) for i in range(0, nrepeat)])
    mape_RF_size = np.mean([mk_treepredictor('binary-size', False, tsize) for i in range(0, nrepeat)])
    
    mape_linear_gadget = np.mean([mk_linear_regression('nr-of-gadgets', tsize) for i in range(0, nrepeat)])
    mape_DT_gadget = np.mean([mk_treepredictor('nr-of-gadgets', True, tsize) for i in range(0, nrepeat)])
    mape_RF_gadget = np.mean([mk_treepredictor('nr-of-gadgets', False, tsize) for i in range(0, nrepeat)])
    
    return {'system' : SYSTEM_NAME, 'corr': corr_size_gadget, 
            'test_size' : tsize, 
            'MAPE_linear_size' : mape_linear_size, 
            'MAPE_DT_size' : mape_DT_size, 
            'MAPE_RF_size' : mape_RF_size,
            'MAPE_linear_gadget' : mape_linear_gadget, 
            'MAPE_DT_gadget' : mape_DT_gadget, 
            'MAPE_RF_gadget' : mape_RF_gadget
           }
    
    
plt.style.use('ggplot')

summary_systems = pd.DataFrame(columns=['system', 'corr', 'test_size', 
                                        'MAPE_linear_size', 'MAPE_DT_size', 'MAPE_RF_size',
                                        'MAPE_linear_gadget', 'MAPE_DT_gadget', 'MAPE_RF_gadget'                                       
                                       ])

tsizes = [0.95, 0.9, 0.8, 0.7, 0.5, 0.1]
for tsize in tsizes:
    xz = process_system(DATASET='xz/splc21-data - xz-200.csv', SYSTEM_NAME='xz', tsize=tsize)
    summary_systems = summary_systems.append(xz, ignore_index=True)
    nginx = process_system(DATASET='nginx/splc21-data - nginx-200.csv', SYSTEM_NAME='nginx', tsize=tsize)
    summary_systems = summary_systems.append(nginx, ignore_index=True)
    x264 = process_system(DATASET='x264/splc21-data - x264-200.csv', SYSTEM_NAME='x264', tsize=tsize)
    summary_systems = summary_systems.append(x264, ignore_index=True)
    sqlite = process_system(DATASET='sqlite/splc21-data - sqlite-200.csv', SYSTEM_NAME='sqlite', tsize=tsize)
    summary_systems = summary_systems.append(sqlite, ignore_index=True)

summary_systems
```
