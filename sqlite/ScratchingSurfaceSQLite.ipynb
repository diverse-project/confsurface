```python
import pandas as pd

DATASET='splc21-data - sqlite-200.csv'
SYSTEM_NAME='sqlite'

FTS=['--disable-option-checking', '--disable-libtool-lock',
       '--disable-largefile', '--disable-threadsafe', '--disable-tcl',
       '--disable-readline', '--disable-amalgamation',
       '--disable-load-extension', '--disable-math', '--enable-shared=no',
       '--enable-static', '--enable-fast-install', '--enable-releasemode',
       '--enable-tempstore', '--enable-editline', '--enable-debug',
       '--enable-all', '--enable-memsys5', '--enable-memsys3', '--enable-fts3',
       '--enable-fts4', '--enable-fts5', '--enable-json1',
       '--enable-update-limit', '--enable-geopoly', '--enable-rtree',
       '--enable-session', '--enable-gcov', '--with-pic', '--with-gnu-ld',
       '--with-tcl=/usr/lib64']

odf = pd.read_csv(DATASET)
odf.head()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn import linear_model
import graphviz 

def mean_absolute_percentage_error(y_pred, y_true):
    return np.average(np.abs(y_pred - y_true) / np.abs(y_true))

def eval_model(target, y_pred, X_test, y_test): 
    print('Mean squared error: %.2f'
          % mean_squared_error(y_test, y_pred))
    # The coefficient of determination: 1 is perfect prediction
    print('Coefficient of determination: %.2f'
          % r2_score(y_test, y_pred))
    
    print("MAPE", mean_absolute_percentage_error(y_pred, y_test))

    # Plot outputs
    plt.scatter(X_test.index, y_test,  color='black')
    plt.plot(X_test.index, y_pred, color='blue', linewidth=3)

    plt.xticks(())
    plt.yticks(())

    plt.show()

def mk_linear_regression(target):
    regr = linear_model.LinearRegression()
    X_train, X_test, y_train, y_test = train_test_split(odf[FTS], odf[target], test_size=0.8, random_state=42)
    regr.fit(X_train, y_train)
    print(regr.coef_)
    coeffs = pd.DataFrame(regr.coef_, FTS, columns=['Coefficients'])
    print(coeffs.sort_values(by="Coefficients"))
    coeffs.plot(kind='barh', figsize=(9, 7))
    plt.axvline(x=0, color='.5')
    plt.subplots_adjust(left=.3)
    plt.show()
    
    y_pred = regr.predict(X_test)
    eval_model(target, y_pred, X_test, y_test)
    



def mk_treepredictor(target, decision_tree): 
    if decision_tree:
        clf = tree.DecisionTreeRegressor()
    else:
        clf = RandomForestRegressor() 
    X_train, X_test, y_train, y_test = train_test_split(odf[FTS], odf[target], test_size=0.8, random_state=42)
   
    clf = clf.fit(X_train, y_train)

    if decision_tree:
        dot_data = tree.export_graphviz(clf, out_file=None, feature_names=FTS,  
                      filled=True, rounded=True,  
                      special_characters=True)
        graph = graphviz.Source(dot_data) 
        graph.render(SYSTEM_NAME)
                   
    
    y_pred = clf.predict(X_test)
    eval_model(target, y_pred, X_test, y_test)
    
    importance = clf.feature_importances_
    for i, v in enumerate(importance):
        print('Feature: %d, %s, Score: %.5f' % (i, odf.columns[i], v))
    plt.bar([x for x in range(len(importance))], importance)
    plt.show()
    
mk_linear_regression('binary-size')
mk_treepredictor('binary-size', True)
mk_treepredictor('binary-size', False)
```

```python
mk_linear_regression('nr-of-gadgets')
mk_treepredictor('nr-of-gadgets', True)
mk_treepredictor('nr-of-gadgets', False)
```
